```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("Images/Assignment 2 Image.jpeg")
exp_image = cv2.imread("Images/Expected Output.jpg")

image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
exp_image = cv2.cvtColor(exp_image, cv2.COLOR_BGR2GRAY)

plt.title("Original Image")
plt.imshow(image, cmap="gray")
plt.show()

plt.title("Expected Image")
plt.imshow(exp_image, cmap="gray")
plt.show()
```

### Image Histogram

```python
hist = cv2.calcHist([image], [0], None, [256], [0, 256])

plt.plot(hist)
```

### Expected Image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

blurred_image = cv2.GaussianBlur(image, (3, 3), 0)

# Low Pass Filter
threshold_value = 210
thresholded_image = np.where(blurred_image < threshold_value, image, 0)

plt.imshow(thresholded_image, cmap="gray")
plt.show()
```
