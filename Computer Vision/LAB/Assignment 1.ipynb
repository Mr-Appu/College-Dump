```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("Images/sunset.jpg")

print("Image Shape : ", image.shape)

plt.title("Original Image")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

### Image Resizing

#### Scale : 0.2

```python
scale = 0.2

new_height = int(scale * image.shape[0])
new_width  = int(scale * image.shape[1])

name = ["NEAREST", "LINEAR", "CUBIC", "INTER AREA"]

for idx, inter_polation in enumerate([cv2.INTER_NEAREST, cv2.INTER_LINEAR, cv2.INTER_CUBIC, cv2.INTER_AREA]):
    new_image = cv2.resize(image, (new_width, new_height), interpolation = inter_polation)
    
    plt.title(name[idx])
    plt.imshow(cv2.cvtColor(new_image, cv2.COLOR_BGR2RGB))
    plt.show()
```

#### Scale : 1.5

```python
scale = 1.5

new_height = int(scale * image.shape[0])
new_width  = int(scale * image.shape[1])

name = ["NEAREST", "LINEAR", "CUBIC", "INTER AREA"]

for idx, inter_polation in enumerate([cv2.INTER_NEAREST, cv2.INTER_LINEAR, cv2.INTER_CUBIC, cv2.INTER_AREA]):
    new_image = cv2.resize(image, (new_width, new_height), interpolation = inter_polation)
    
    plt.title(name[idx])
    plt.imshow(cv2.cvtColor(new_image, cv2.COLOR_BGR2RGB))
    plt.show()
```

### Image Type Conversion

```python
# jpg to png
cv2.imwrite("Images/sunset.png", image)

# jpg to jpeg
cv2.imwrite("Images/sunset.jpeg", image)

# Read PNG File
png_image = cv2.imread("Images/sunset.png")

plt.title("PNG Image")
plt.imshow(cv2.cvtColor(png_image, cv2.COLOR_BGR2RGB))
plt.show()
```

### Extraction of color bands from an image

```python
blue_channel, green_channel, red_channel = cv2.split(image)

title = ["Red Channel", "Green Channel", "Blue Channel"]
img_l = [red_channel, green_channel, blue_channel]
cmap_l = ["Reds", "Greens", "Blues"]

for idx in range(3):
    
    plt.title(title[idx])
    
    plt.imshow(img_l[idx], cmap=cmap_l[idx])
    plt.show()
```
